# Comparison charts

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Column chart

**Create Data**

```python
# choose columns
column1 = [0.1, 17.5, 40, 48, 52, 69, 88]
column2 = [2, 8, 70, 1.5, 25, 12, 28]

# name to display for every column on y-axis
index = ['snail', 'pig', 'elephant','rabbit', 'giraffe', 'coyote', 'horse']


# create a Pandas dataframe with this info
df = pd.DataFrame({
    'Name': index,
    'Height(in cm)': column1,
    'Age': column2
})
```

```python
# MATPLOTLIB ( built into python)

# plotting graph
df.plot(x="Name", y=["Age", "Height(in cm)"], kind="bar")

# Show the plot
plt.show()
```

```python
# PANDAS 

# plot it
ax = df.plot.bar(x="Name", y=["Age", "Height(in cm)"], rot=0)
```

```python
# SEABORN

df.plot(kind="bar", x="Name", y=["Age", "Height(in cm)"], figsize=(7, 5))

# add label text
plt.xlabel("Name")
```

```python
# PLOTLY

fig = px.bar(df, 
             x="Name", 
             y=["Age", "Height(in cm)"], 
             barmode='group', 
             color_discrete_map={
                 'Age': 'blue',
                 'Height(in cm)': 'red'},
             width= 50  # Change bin width
            )

fig.update_layout(
    width=800,  # Change figure width
    height=500  # Change figure height
)

fig.show()
```
