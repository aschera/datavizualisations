# Comparison charts

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
````

````python
# choose columns
column1 = [0.1, 17.5, 40, 48, 52, 69, 88]
column2 = [0.5, 19, 35, 53, 57, 72, 89]

# name to display for every column
index = ['snail', 'pig', 'elephant','rabbit', 'giraffe', 'coyote', 'horse']


# create a Pandas dataframe with this info
df = pd.DataFrame({
    'name': index,
    'height': column1,
    'age': column2
})
````

Both the Bar and the Column charts display data using rectangular bars where the length of the bar is proportional to the data value. 

## 1 Bar chart

A bar chart is a type of chart that is often used to display categorical data with numerical values. 
In a bar chart, the **categories are typically displayed on the vertical axis**, while the **numerical values are displayed on the horizontal axis**. 
However, in some cases, it can be useful to display the categories on the horizontal axis instead. 
This type of chart is called a horizontal bar chart.

Bar charts are particularly suitable for displaying long data labels, as they provide ample space to display text without cluttering the chart. 
In contrast, column charts have limited space in the category axis, which can make the chart look cluttered when the data labels are long.

When you need to display a large number of data sets on the category axis, a bar chart can be a good choice. 
In this case, it is often best to place the data sets one below the other, which makes optimal use of space and ensures that the chart remains easy to read.

### 1.1 MATPLOTLIB 

````python
# MATPLOTLIB ( built into python)

# plotting graph
ax = df.plot(x="name", y=["age", "height"], kind="barh")

# Move legend to the right side
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.92))

# Show the plot
plt.show()
````

```
kind="barh"
```
This makes the bars go on the horizontal axis ( "bar" makes them go on the vertical axes).

### 1.2 PANDAS

````python
# PANDAS 

# plot it
ax = df.plot.barh(x="name", y=["age", "height"], legend=True)

# Move legend to the right side
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.94))

# Show the plot
plt.show()
````

### 1.3 SEABORN

````python
# SEABORN

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

# melt the DataFrame to create a tidy format
df_melt = pd.melt(df, id_vars=['name'], value_vars=['age', 'height'])

# create the plot
plt.figure(figsize=(7, 5))
sns.barplot(x='value', y='name', hue='variable', data=df_melt, palette=['blue', 'orange'],  order=df['name'][::-1])
plt.xlabel('Value')
plt.ylabel('name')
plt.title('age and height')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
plt.show()
````

### 1.4 PLOTLY

````python
# PLOTLY

fig = px.bar(df, 
             x=["age", "height"], 
             y="name", 
             barmode='group', 
             color_discrete_sequence=['blue', 'darkorange'],
             orientation='h',
             width=60  # Change bin width
            )

# Rotate x-axis labels
fig.update_layout(xaxis={'tickangle': 90})

# Change figure width and height
fig.update_layout(
    width=700,  # Change figure width
    height=500  # Change figure height
)

fig.update_layout(boxgroupgap=0.5, boxgap=0.1)

fig.update_layout(legend_valign='top')

# Show the plot
fig.show()
````

## 2 Column chart

A column chart is a type of chart that is often used to display categorical data with numerical values. 
In a column chart, the categories are typically displayed on the horizontal axis, while the numerical values are displayed on the vertical axis. This type of chart is best suited for displaying up to 10-12 data sets.

However, when you need to display a large number of data sets, using a column chart is not the best way forward. 
This is because the limited space in the category axis can make the chart look cluttered and difficult to read. In such cases, a different chart type, such as a bar chart, may be more appropriate.

One of the advantages of using a column chart is that it is suitable for displaying data sets with negative values. 
Since we associate negative values with the downward direction, using a column chart allows us to display such values in a way that is consistent with our mental model of vertical space.

## Grouped column chart

**Create Data**

````python
# MATPLOTLIB ( built into python)

# plotting graph
df.plot(x="Name", y=["Age", "Height(in cm)"], kind="bar")

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.show()
````

````python
# PANDAS 

# plot it
ax = df.plot.bar(x="Name", y=["Age", "Height(in cm)"], rot=0)

# Rotate x-axis labels
ax.tick_params(axis='x', labelrotation=90)
````

````python
# SEABORN

df.plot(kind="bar", x="Name", y=["Age", "Height(in cm)"], figsize=(7, 5))

# add label text
plt.xlabel("Name")
````

````python
# PLOTLY

fig = px.bar(df, 
             x="Name", 
             y=["Age", "Height(in cm)"], 
             barmode='group', 
             color_discrete_sequence=['blue', 'darkorange'],
             width=60  # Change bin width
            )

# Rotate x-axis labels
fig.update_layout(xaxis={'tickangle': 90})

# Change figure width and height
fig.update_layout(
    width=700,  # Change figure width
    height=500  # Change figure height
)

fig.update_layout(boxgroupgap=0.5, boxgap=0.1)

fig.update_layout(legend_valign='top')

# Show the plot
fig.show()
````
